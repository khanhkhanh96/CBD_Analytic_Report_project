```python
#lấy mẫu doanh thu của các bộ phim theo ngày xuất bản trong tuần
dataweek = pd.DataFrame(columns=['0','1','2','3','4','5','6'])
myarr1 = np.zeros((40,7))      
for i in range(7):
    count = 0
    curr = 0
    while count < 40 and curr<4803:
        try:
            idate=datetime.datetime.strptime(data.release_date[curr], '%Y-%m-%d')
            if idate.weekday()==i:
                myarr1[count][i] = data.revenue[curr]
                count+=1
            curr+=1
        except:
            continue
for arr in myarr1:
    dataweek = dataweek.append(pd.Series(arr), ignore_index=True)
dataweek.head()
f,p=scipy.stats.f_oneway(dataweek[0],dataweek[1],dataweek[2],dataweek[3],dataweek[4],dataweek[5],dataweek[6])
fcrit=scipy.stats.f.ppf(q=1-0.05, dfn=6, dfd=153)
print(f,fcrit,p)
```

    Thực hiện ANOVA giữa các ngày trong tuần cho thấy có sự sai khác nhau trong doanh thu của phim có ngày xuất bản trong tuần khác nhau.

```python
revenuewday=np.zeros((7))
countwday=np.zeros((7))
tb=np.zeros((7))
for i in range(7):
    for j in range(4803):
        try:
            idate=datetime.datetime.strptime(data.release_date[j], '%Y-%m-%d')
            if idate.weekday()==i:
                revenuewday[i]+=data.revenue[j]
                countwday[i]+=1
        except:
            continue
    tb[i]=revenuewday[i]/countwday[i]
weekdata = pd.DataFrame({
 'revenue':revenuewday
}, index=list('0123456'))
ax = weekdata.plot(kind='bar',title="Tổng doanh thu theo ngày trong tuần")
plt.show()
```

- Tổng doanh thu tập trung các phim phát hành vào giữa tuần.

```python
weekdata = pd.DataFrame({
 'count':countwday
}, index=list('0123456'))
ax = weekdata.plot(kind='bar',title="số lượng phim công chiếu theo ngày trong tuần")
plt.show()
```
